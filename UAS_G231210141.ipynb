# Histogram Matplotlib

### Import Library, Import CSV File dan Tampilkan Data CSV



```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
myData = pd.read_csv('stock.csv')
# Menampilkan Data File CSV
myData
```

### Distribusi Purchase Price Dengan Histogram matplotlib

```python
# Membuat variabel purchase_price dengan mengambil data dari kolom Purchase Price
purchase_prices = myData['Purchase Price']

# Membuat histogram
plt.hist(purchase_prices, bins=10,color='#746AB0', edgecolor='black')

# Menambahkan judul dan label sumbu
plt.title('Distribusi Purchase Price')
plt.xlabel('Purchase Price')
plt.ylabel('Jumlah')

# Menampilkan histogram
plt.show()
```

### Distribusi Current Price Dengan Histogram matplotlib

```python
# Membuat variabel current_prices dengan mengambil data dari kolom Current Price
current_prices = myData['Current Price']

# Membuat histogram
plt.hist(current_prices, bins=10,color='#288BA8', edgecolor='black')

# Menambahkan judul dan label sumbu
plt.title('Distribusi Current Price')
plt.xlabel('Current Price')
plt.ylabel('Jumlah')

# Menampilkan histogram
plt.show()
```

### Distribusi Gain/Loss Dengan Histogram matplotlib

```python
# membuat variabel gain_loss dengan mengambil data dari kolom Gain/Loss
gain_loss = myData['Gain/Loss']

# Membuat histogram
plt.hist(gain_loss, bins=10,color='#E389B9', edgecolor='black')

# Menambahkan judul dan label sumbu
plt.title('Distribusi Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Jumlah')

# Menampilkan histogram
plt.show()
```

# Histogram Seaborn

### Distribusi Current Price Dengan Histogram Seaborn

```python
# Import library seaborn
import seaborn as sns

# Mengatur gaya seaborn
sns.set(style="darkgrid")

# Membuat histogram dengan seaborn
sns.histplot(data=current_prices, bins=10, kde=True, color='#148F77', edgecolor='black')
```

### Distribusi Purchase Price Dengan Histogram Seaborn

```python
sns.histplot(data=purchase_prices, bins=10, kde=True, color='#6C3483', edgecolor='black')
```

### Top 5 Stock Dengan Dividend/Share Tertinggi

```python
# Mengurutkan data berdasarkan kolom "Dividend/share" secara menurun/descending
sorted_myData = myData.sort_values(by='Dividend/share', ascending=False)

# Mengambil lima data teratas
big5DividendShare = sorted_myData.head(5)

# Menyiapkan label sumbu X dengan format "Stock Name (Symbol)"
label = [f"{name} ({symbol})" for name, symbol in zip(big5DividendShare['Stock Name'], big5DividendShare['Symbol'])]

# Mengambil nilai dividen per saham
dividends = big5DividendShare['Dividend/share']

# Membuat diagram batang (bar chart)
plt.bar(label, dividends, color='#F08080', edgecolor='black')
plt.xlabel('Stock Name (Symbol)')
plt.ylabel('Dividend/Share')
plt.title('Top 5 Stock Dengan Dividend/Share Tertinggi')
plt.xticks(rotation=45)

# Menampilkan diagram batang
plt.show()
```

### Top 5 Stock Dengan Market Value Tertinggi

```python
# Mengurutkan data berdasarkan kolom "Market Value" secara menurun
sorted_myData = myData.sort_values(by='Market Value', ascending=False)

# Mengambil lima data teratas
big5MarketValue = sorted_myData.head(5)

# Menyiapkan label sumbu X dengan format "Stock Name (Symbol)"
label = [f"{name} ({symbol})" for name, symbol in zip(big5MarketValue['Stock Name'], big5MarketValue['Symbol'])]

# Mengambil nilai Market Value
market_values = big5MarketValue['Market Value']

# Membuat diagram batang (bar chart)
plt.bar(label, market_values, color='#17A589', edgecolor='black')
plt.xlabel('Stock Name (Symbol)')
plt.ylabel('Market Value')
plt.title('Top 5 Stock Dengan Market Value Tertinggi')
plt.xticks(rotation=45)

# Menampilkan diagram batang
plt.show()
```

### Top 7 Stock Dengan Annual Yield Tertinggi

```python
# Mengurutkan data berdasarkan kolom "Annual Yield" secara descending
sorted_myData = myData.sort_values(by='Annual Yield', ascending=False)

# Mengambil tujuh data teratas
big7AnnualYield = sorted_myData.head(7)

# Menyiapkan label sumbu X dengan format "Stock Name (Symbol)"
label = [f"{name} ({symbol})" for name, symbol in zip(big7AnnualYield['Stock Name'], big7AnnualYield['Symbol'])]

# Mengambil nilai Annual Yield
annual_yields = big7AnnualYield['Annual Yield']

# Membuat diagram batang (bar chart)
plt.bar(label, annual_yields, color='#8b64b4', edgecolor='black')
plt.xlabel('Stock Name (Symbol)', color='black')
plt.ylabel('Annual Yield')
plt.title('Top 7 Stock Dengan Annual Yield Tertinggi')
plt.xticks(rotation=90)

# Menampilkan diagram batang
plt.show()
```
